```python
# Importa as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

# ----- Dados do Exercício -----
# Hidrograma Unitário (HU) para uma chuva efetiva de 10 mm com duração de 1h
tempo_hu = np.array([0, 1, 2, 3, 4, 5, 6])
vazao_hu_10mm = np.array([0, 45, 70, 50, 25, 10, 0])

# Chuvas efetivas a serem utilizadas para a previsão do hidrograma de cheia
chuva_1_mm = 30  # Primeira chuva efetiva de 30 mm
chuva_2_mm = 20  # Segunda chuva efetiva de 20 mm

# Duração de cada chuva (em horas)
duracao_chuva = 1  # 1h

# ----- Processamento dos Dados -----
# 1. Normalização do HU para 1 mm (princípio da linearidade)
vazao_hu_1mm = vazao_hu_10mm / 10.0

# 2. Geração dos hidrogramas individuais a partir das chuvas efetivas
hidrograma_chuva_1 = vazao_hu_1mm * chuva_1_mm
hidrograma_chuva_2 = vazao_hu_1mm * chuva_2_mm

# 3. Convolução: Soma dos hidrogramas deslocados no tempo
# O hidrograma total terá um tamanho maior devido ao deslocamento
tamanho_total = len(hidrograma_chuva_1) + duracao_chuva

# Cria arrays com zeros para armazenar os hidrogramas individuais e o total
hidrograma_1_completo = np.zeros(tamanho_total)
hidrograma_2_deslocado = np.zeros(tamanho_total)
tempo_total = np.arange(tamanho_total)

# Preenche os arrays com os dados dos hidrogramas
hidrograma_1_completo[:len(hidrograma_chuva_1)] = hidrograma_chuva_1
hidrograma_2_deslocado[duracao_chuva:duracao_chuva + len(hidrograma_chuva_2)] = hidrograma_chuva_2

# Hidrograma total de cheia é a soma dos hidrogramas deslocados
hidrograma_total = hidrograma_1_completo + hidrograma_2_deslocado

# 4. Análise do Hidrograma Total
vazao_maxima = np.max(hidrograma_total)
posicao_maxima = np.argmax(hidrograma_total)
tempo_maxima = tempo_total[posicao_maxima]

# ----- Impressão dos Resultados Numéricos -----
print("=== Resultados Numéricos ===")
print(f"HU Normalizado para 1 mm:\n{vazao_hu_1mm}")
print("-" * 40)
print(f"Hidrograma da 1ª chuva ({chuva_1_mm} mm):\n{hidrograma_1_completo}")
print(f"Hidrograma da 2ª chuva ({chuva_2_mm} mm) deslocado:\n{hidrograma_2_deslocado}")
print("-" * 40)
print(f"Hidrograma de Cheia (soma):\n{hidrograma_total}")
print("-" * 40)
print(f"A vazão máxima de escoamento superficial direto é: {vazao_maxima} m³/s")
print(f"Ocorre no tempo: {tempo_maxima} h")

# ----- Geração do Gráfico -----
plt.style.use('seaborn-v0_8-whitegrid') # Define um estilo para o gráfico

plt.figure(figsize=(12, 7))

# Plota os hidrogramas individuais com linhas tracejadas
plt.plot(tempo_total, hidrograma_1_completo, '--', label=f'Hidrograma da 1ª Chuva ({chuva_1_mm} mm)', color='blue', alpha=0.7)
plt.plot(tempo_total, hidrograma_2_deslocado, '--', label=f'Hidrograma da 2ª Chuva ({chuva_2_mm} mm)', color='green', alpha=0.7)

# Plota o hidrograma total (a soma) com uma linha mais espessa
plt.plot(tempo_total, hidrograma_total, '-', label='Hidrograma de Cheia Total', color='red', linewidth=3)

# Adiciona o ponto de vazão máxima
plt.plot(tempo_maxima, vazao_maxima, 'ro', markersize=10)
plt.text(tempo_maxima + 0.1, vazao_maxima, f'{vazao_maxima} m³/s', fontsize=12, verticalalignment='bottom')

# Adiciona títulos e rótulos
plt.title('Hidrograma de Cheia Resultante da Convolução', fontsize=16)
plt.xlabel('Tempo (h)', fontsize=12)
plt.ylabel('Vazão (m³/s)', fontsize=12)
plt.legend(loc='upper right', fontsize=10)
plt.xticks(np.arange(0, tamanho_total, 1)) # Garante que os rótulos do eixo X sejam inteiros

# Adiciona grade
plt.grid(True)

# Mostra o gráfico
plt.show()
```

```python
import numpy as np
# Chuvas efetivas (mm)
i = np.array([7.0, 2.0, 1.3, 0.5, 1.4, 2.0, 2.1, 0.5, 0.5, 0.5, 0.5])
# Vazões do escoamento superficial direto (m³/s)
Q_ESD = np.array([0, 249, 3994, 6151, 8386, 11424, 13449, 15014, 16036, 16837, 17403, 17849, 18042, 17592, 17019, 16452, 15644, 14729, 13599, 12274, 10673, 8946, 7228, 5803, 4636, 3541, 2735, 1973, 1448, 1068, 582, 287, 0])

# Duração do evento de chuva
n_chuvas = len(i)
# Duração do hidrograma
n_Q = len(Q_ESD)
# O comprimento do HU é n_Q - n_chuvas + 1
n_HU = n_Q - n_chuvas + 1
HU = np.zeros(n_HU)

# Algoritmo de deconvolução
# A equação para cada ponto de tempo Q_j é:
# Q_j = i_1*HU_j + i_2*HU_{j-1} + ... + i_m*HU_{j-m+1}

# Para j=0, Q_0 = i_0 * HU_0 => 0 = 7.0 * HU_0 => HU_0 = 0
HU[0] = Q_ESD[0] / i[0] if i[0] != 0 else 0

# Loop para calcular cada valor do hidrograma unitário
for j in range(1, n_HU):
    # Calcular a soma dos termos de convolução já conhecidos
    soma_conhecida = 0
    # O loop interno vai de 1 até min(j, n_chuvas - 1)
    for k in range(1, min(j + 1, n_chuvas)):
        soma_conhecida += i[k] * HU[j-k]

    # A equação para o ponto j é: Q_j = i[0]*HU[j] + soma_conhecida
    # Reorganizando: HU[j] = (Q_j - soma_conhecida) / i[0]
    HU[j] = (Q_ESD[j] - soma_conhecida) / i[0]

# Imprimir os resultados
print("Hidrograma Unitário (m³/s por mm):")
print("-" * 40)
for k in range(n_HU):
    tempo = k * 30  # minutos
    print(f"Tempo = {tempo} min:  HU = {HU[k]:.4f} m³/s")

# A área do HU (m³/s * 30 min) deve ser igual ao volume da chuva
# 30 minutos = 30 * 60 = 1800 segundos
# A área do hidrograma unitário é a soma dos valores de vazão multiplicada pelo intervalo de tempo (1800 s)
volume_HU = np.sum(HU) * 1800 # m³
print("-" * 40)
print(f"Volume do HU (m³): {volume_HU:.2f}")

# A área da bacia (A_bacia) em m² seria volume_HU / (1 mm * 1m/1000mm)
# A_bacia = volume_HU / 0.001
# print(f"Área da bacia (km²): {A_bacia / 1e6:.2f}")
```
