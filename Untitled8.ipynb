```python
from ipyleaflet import Map, GeoData, Marker, LayersControl
from ipywidgets import HTML, Output
import geopandas as gpd

# Centro do mapa
center = [streams_wgs.geometry.centroid.y.mean(), streams_wgs.geometry.centroid.x.mean()]

# Cria o mapa
m = Map(center=center, zoom=12)

# Adiciona a rede de drenagem
streams_layer = GeoData(geo_dataframe=streams_wgs, name='Rede de Drenagem')
m.add_layer(streams_layer)

# Controle de camadas
m.add_control(LayersControl())

# Widget para saída
out = Output()
display(out)

# Variável global para armazenar o marcador
marker = None

# Função que captura clique no mapa
def handle_map_click(**kwargs):
    global marker
    if kwargs.get('type') == 'click':
        latlon = kwargs.get('coordinates')
        with out:
            print(f"Ponto clicado: {latlon}")

        # Remove marcador anterior, se existir
        if marker:
            m.remove_layer(marker)

        # Cria novo marcador no local clicado
        marker = Marker(location=latlon, draggable=True)
        m.add_layer(marker)

# Ativa a função de clique
m.on_interaction(handle_map_click)

m
```

Retortar a rede de drenagem para dentro da bacia apenas:

```python
import geopandas as gpd
import rasterio
from rasterio.mask import mask

# Carregar shapefile da bacia
bacia = gpd.read_file('/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/bacia.shp')
bacia = bacia.to_crs(epsg=31982)  # CRS compatível com os rasters

# Função para recortar raster
def clip_raster(raster_path, shapefile, output_path):
    with rasterio.open(raster_path) as src:
        out_image, out_transform = mask(src, shapefile.geometry, crop=True)
        out_meta = src.meta.copy()

        out_meta.update({
            "driver": "GTiff",
            "height": out_image.shape[1],
            "width": out_image.shape[2],
            "transform": out_transform
        })

        with rasterio.open(output_path, "w", **out_meta) as dest:
            dest.write(out_image)

# Recorta direção de fluxo
clip_raster(
    '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/flow_dir.tif',
    bacia,
    '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/flow_dir_clip.tif'
)

# Recorta rede de drenagem
clip_raster(
    '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/streams.tif',
    bacia,
    '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/streams_clip.tif'
)

```

aqui converte o raster para shape:

```python
import rasterio
from rasterio import features
import geopandas as gpd
from shapely.geometry import shape

# Caminho do arquivo raster
raster_path = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/ordem_strahler_clip.tif'

# Abrir raster
with rasterio.open(raster_path) as src:
    image = src.read(1)  # ler banda 1 (ordem)
    mask = image > 0     # máscara para ignorar zeros (sem dados)
    transform = src.transform
    crs = src.crs

# Extrair as geometrias (polígonos) e os valores dos pixels
shapes = features.shapes(image, mask=mask, transform=transform)

# Criar lista de dicionários para montar GeoDataFrame
records = []
for geom, value in shapes:
    records.append({
        'geometry': shape(geom),
        'ordem': int(value)
    })

# Criar GeoDataFrame
gdf = gpd.GeoDataFrame(records, crs=crs)

# Opcional: filtrar só polígonos, caso existam outros tipos geométricos
gdf = gdf[gdf.geometry.type == 'Polygon']

# Salvar shapefile
output_shp = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/ordem_strahler_clip.shp'
gdf.to_file(output_shp)

print(f'Shapefile salvo em: {output_shp}')
```

```python
import rasterio
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from rasterio.features import rasterize
from rasterio.mask import mask

# --- 1. Tabela de CN ---
# Mapeamento de uso do solo (MapBiomas) + Grupo Hidrológico → CN
cn_lookup_table = {
    # 0 - Sem dados / áreas não classificadas
    (0, 'A'): -9999, (0, 'B'): -9999, (0, 'C'): -9999, (0, 'D'): -9999,

    # 3 - Floresta
    (3, 'A'): 36, (3, 'B'): 60, (3, 'C'): 70, (3, 'D'): 76,

    # 9 - Formação Campestre
    (9, 'A'): 49, (9, 'B'): 69, (9, 'C'): 79, (9, 'D'): 84,

    # 11 - Outras Formações não Florestais
    (11, 'A'): 45, (11, 'B'): 66, (11, 'C'): 77, (11, 'D'): 83,

    # 12 - Formação Campestre (possível duplicata de 9, mantido valor típico)
    (12, 'A'): 49, (12, 'B'): 69, (12, 'C'): 79, (12, 'D'): 84,

    # 15 - Pastagem
    (15, 'A'): 25, (15, 'B'): 59, (15, 'C'): 75, (15, 'D'): 83,

    # 19 - Agricultura (Mosaico)
    (19, 'A'): 68, (19, 'B'): 79, (19, 'C'): 84, (19, 'D'): 88,

    # 21 - Área Urbana
    (21, 'A'): 77, (21, 'B'): 85, (21, 'C'): 90, (21, 'D'): 92,

    # 24 - Mineração
    (24, 'A'): 77, (24, 'B'): 86, (24, 'C'): 91, (24, 'D'): 94,

    # 25 - Outras Áreas não Vegetadas
    (25, 'A'): 72, (25, 'B'): 82, (25, 'C'): 88, (25, 'D'): 91,

    # 33 - Corpos d'Água
    (33, 'A'): 100, (33, 'B'): 100, (33, 'C'): 100, (33, 'D'): 100
}


# --- 2. Mapeamento nc_ordem → Grupo Hidrológico (exemplo, preencher conforme PDF) ---
mapa_grupo_hidro = {
    "AFLORAMENTO ROCHOSO": "D",
    "ARGISSOLO": "C",
    "CAMBISSOLO": "C",
    "CHERNOSSOLO": "B",
    "DUNAS": "A",
    "ESPODOSSOLO": "A",
    "GLEISSOLO": "D",
    "ILHA": "D",  # pode ser ajustado caso tenha mais detalhe
    "LATOSSOLO": "B",
    "LUVISSOLO": "C",
    "NEOSSOLO": "A",
    "NITOSSOLO": "B",
    "ORGANOSSOLO": "D",
    "PLANOSSOLO": "D",
    "PLINTOSSOLO": "D",
    "SOLO DE MANGUE": "D",
    "SOLO HIDROMÓRFICO": "D",
    "VERTISSOLO": "D"
}


# --- 3. Caminhos ---
caminho_bacia = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/bacia.shp"
caminho_mapa_usosolo = "/content/drive/MyDrive/hidrologia/mapbiomas_uso_bacia_2022.tif"
caminho_mapa_solos_shp = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/pedologia/pedologia_consolidada.shp"
caminho_mapa_cn_saida = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/mapa_cn.tif"

# --- 4. Função ---
def estimar_cn(mapa_usosolo_tif, mapa_solos_shp, bacia_shp, cn_tabela, mapa_cn_saida_tif):
    # Ler bacia
    bacia = gpd.read_file(bacia_shp)
    bacia = bacia.to_crs("EPSG:4326")

    # --- Recorte do raster de uso do solo ---
    with rasterio.open(mapa_usosolo_tif) as src:
        bacia_proj = bacia.to_crs(src.crs)
        uso_clip, transform_clip = mask(src, bacia_proj.geometry, crop=True)
        uso_clip = uso_clip[0]
        out_meta = src.meta.copy()
        out_meta.update({
            "height": uso_clip.shape[0],
            "width": uso_clip.shape[1],
            "transform": transform_clip
        })
        crs_uso = src.crs
        height, width = uso_clip.shape

    # --- Recorte do shapefile de solos ---
    solos_gdf = gpd.read_file(mapa_solos_shp)
    solos_gdf = solos_gdf.to_crs(crs_uso)
    solos_clip = gpd.clip(solos_gdf, bacia_proj)

    # Mapear grupo hidrológico
    if 'nc_ordem' not in solos_clip.columns:
        raise ValueError(f"Coluna 'nc_ordem' não encontrada. Colunas: {solos_clip.columns.tolist()}")
    solos_clip['grupo_hidro'] = solos_clip['nc_ordem'].map(mapa_grupo_hidro)

    # Rasterizar solos
    grupo_val_map = {'A': 1, 'B': 2, 'C': 3, 'D': 4}
    solos_clip['grupo_val'] = solos_clip['grupo_hidro'].map(grupo_val_map)
    shapes_raster = [(geom, val) for geom, val in zip(solos_clip.geometry, solos_clip['grupo_val'])]
    solo_array = rasterize(
        shapes=shapes_raster,
        out_shape=(height, width),
        transform=transform_clip,
        fill=0,
        dtype=np.uint8
    )
    inv_grupo_val = {v: k for k, v in grupo_val_map.items()}
    solo_str = np.vectorize(inv_grupo_val.get)(solo_array)

    # Criar CN array
    cn_array = np.full((height, width), -9999.0, dtype=np.float32)
    for i in range(height):
        for j in range(width):
            uso_code = uso_clip[i, j]
            grupo = solo_str[i, j]
            if grupo in ['A', 'B', 'C', 'D']:
                cn_array[i, j] = cn_tabela.get((uso_code, grupo), -9999.0)

    # Salvar raster CN
    out_meta.update({"dtype": "float32", "nodata": -9999.0})
    with rasterio.open(mapa_cn_saida_tif, 'w', **out_meta) as dest:
        dest.write(cn_array, 1)

    # CN médio
    cn_validos = cn_array[cn_array != -9999.0]
    if cn_validos.size > 0:
        print(f"✅ CN médio: {cn_validos.mean():.2f}")
    print(f"✅ Mapa salvo em: {mapa_cn_saida_tif}")

    # --- Visualização ---
    plt.figure(figsize=(10, 8))
    img = plt.imshow(np.where(cn_array == -9999.0, np.nan, cn_array),
                     cmap='Spectral_r',
                     extent=[
                         transform_clip[2],
                         transform_clip[2] + width * transform_clip[0],
                         transform_clip[5] + height * transform_clip[4],
                         transform_clip[5]
                     ])
    plt.colorbar(img, label="Curve Number (CN)")
    bacia.boundary.plot(ax=plt.gca(), color='black', linewidth=1.5)
    plt.title("Mapa de Curve Number (CN) - Recortado para a Bacia")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.show()

# --- 5. Executar ---
estimar_cn(caminho_mapa_usosolo, caminho_mapa_solos_shp, caminho_bacia, cn_lookup_table, caminho_mapa_cn_saida)
```

**Fator de Forma (Ff)**

O Fator de Forma compara a área da bacia com o quadrado do seu comprimento axial máximo. Quanto menor, mais alongada a bacia.

Fórmula: Ff = A / L^2
 Onde A é a área em km² e L é o comprimento axial em km.

```python
import geopandas as gpd
from shapely.geometry import LineString, Point
from shapely.ops import nearest_points
import matplotlib.pyplot as plt
import numpy as np

# === Função para interpretar Kf ===
def interpretar_fator_forma(kf):
    if kf > 0.5:
        return "Bacia quase circular\n(resposta rápida)"
    elif 0.3 <= kf <= 0.5:
        return "Bacia levemente alongada\n(resposta moderada)"
    else:
        return "Bacia muito alongada\n(resposta lenta)"

# === 1. Leitura dos dados ===
bacia_gdf = gpd.read_file('/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/bacia.shp')
exutorio_gdf = gpd.read_file('/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/meu_ponto.shp')

# === 2. Corrige MultiPolygon e mantém o maior ===
bacia_unida = bacia_gdf.geometry.union_all()
bacia_polygon = max(bacia_unida.geoms, key=lambda g: g.area) if bacia_unida.geom_type == 'MultiPolygon' else bacia_unida

exutorio = exutorio_gdf.geometry.iloc[0]

# === 3. Calcula ponto mais distante da bacia ===
coords = np.array(bacia_polygon.exterior.coords)
pontos_borda = [Point(x, y) for x, y in coords]
mais_distante = max(pontos_borda, key=lambda p: p.distance(exutorio))

linha_axial = LineString([exutorio, mais_distante])
comprimento_axial = linha_axial.length

# === 4. Gera pontos ao longo da linha axial ===
n_secoes = 20
pontos_axiais = [linha_axial.interpolate(i / (n_secoes - 1), normalized=True) for i in range(n_secoes)]

# === 5. Construção das seções ortogonais completas ===
larguras = []
secoes = []

dx = linha_axial.coords[-1][0] - linha_axial.coords[0][0]
dy = linha_axial.coords[-1][1] - linha_axial.coords[0][1]
comprimento = np.hypot(dx, dy)
ux, uy = dx / comprimento, dy / comprimento
ort_x, ort_y = -uy, ux

# Estende as seções até garantir cruzamento total com a bacia
offset_inicial = bacia_polygon.bounds[2] - bacia_polygon.bounds[0]  # largura da bacia em X
offset = offset_inicial * 1.5  # segurança adicional

for ponto in pontos_axiais:
    p1 = Point(ponto.x + ort_x * offset, ponto.y + ort_y * offset)
    p2 = Point(ponto.x - ort_x * offset, ponto.y - ort_y * offset)
    secao = LineString([p1, p2])

    intersecao = bacia_polygon.intersection(secao)

    if intersecao.is_empty:
        continue

    if intersecao.geom_type == 'MultiLineString':
        linhas = list(intersecao.geoms)
        maior = max(linhas, key=lambda l: l.length)
        pontos = [Point(maior.coords[0]), Point(maior.coords[-1])]
    elif intersecao.geom_type == 'LineString':
        pontos = [Point(intersecao.coords[0]), Point(intersecao.coords[-1])]
    else:
        continue

    if len(pontos) >= 2:
        largura = pontos[0].distance(pontos[-1])
        larguras.append(largura)
        secoes.append(LineString([pontos[0], pontos[-1]]))

# === 6. Cálculo do fator de forma ===
largura_media = np.mean(larguras)
kf = largura_media / comprimento_axial
interpretacao = interpretar_fator_forma(kf)

# === 7. Plotagem ===
fig, ax = plt.subplots(figsize=(10, 10))
gpd.GeoSeries([bacia_polygon]).plot(ax=ax, color='lightblue', edgecolor='black')
gpd.GeoSeries([linha_axial]).plot(ax=ax, color='red', linewidth=2, label='Comprimento Axial')

if secoes:
    gpd.GeoSeries(secoes).plot(ax=ax, color='green', linewidth=1.5, linestyle='--', label='Seções Ortogonais')
else:
    print("⚠️ Nenhuma seção ortogonal foi gerada.")

# Anotações
ax.annotate(f'L̄ = {largura_media/1000:.2f} km',
            xy=linha_axial.interpolate(0.25, normalized=True).coords[0],
            color='green', fontsize=10)
ax.annotate(f'Laxial = {comprimento_axial/1000:.2f} km',
            xy=linha_axial.interpolate(0.75, normalized=True).coords[0],
            color='red', fontsize=10)

# Título com interpretação
ax.set_title(f'Fator de Forma (Kf) = {kf:.2f}\n{interpretacao}', fontsize=14)
ax.set_axis_off()
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from shapely.geometry import Polygon, box
from sklearn.decomposition import PCA

# --- 1. CARREGAR SHAPE DA BACIA ---
path_bacia_shp = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/bacia.shp'
bacia = gpd.read_file(path_bacia_shp)

# --- 2. Garantir CRS métrico ---
if bacia.crs is None:
    bacia.set_crs(epsg=31982, inplace=True)
else:
    bacia = bacia.to_crs(epsg=31982)

# Corrigir geometria e limpar
bacia['geometry'] = bacia['geometry'].buffer(0)
bacia = bacia[bacia.geometry.type.isin(['Polygon', 'MultiPolygon'])]
bacia = bacia[~bacia.geometry.is_empty & ~bacia.geometry.isna()]

# --- 3. Cálculo de métricas ---
bacia['area_m2'] = bacia.geometry.area
bacia['perimetro_m'] = bacia.geometry.length
area = bacia['area_m2'].sum()
perimetro = bacia['perimetro_m'].sum()

area_km2 = area / 1e6
perimetro_km = perimetro / 1e3

# Coeficiente de Compacidade
Kc = perimetro / (2 * np.sqrt(np.pi * area))

# Comprimento principal da bacia (em km)
comprimento_km = 45.2
Ic = area_km2 / (comprimento_km ** 2)

# --- Função para classificar bacia ---
def classificacao_ic(Ic, limiar=0.5):
    if Ic < limiar:
        return "Bacia alongada"
    else:
        return "Bacia não alongada"

classificacao = classificacao_ic(Ic)

# --- 4. PCA e elipse ---
coords_list = []
for geom in bacia.geometry:
    if geom.geom_type == 'Polygon':
        coords_list.extend(np.array(geom.exterior.coords))
    elif geom.geom_type == 'MultiPolygon':
        for part in geom.geoms:
            coords_list.extend(np.array(part.exterior.coords))

all_coords = np.array(coords_list)

pca = PCA(n_components=2)
pca.fit(all_coords)

center = pca.mean_
components = pca.components_

coords_pca = pca.transform(all_coords)
min_pca = coords_pca.min(axis=0)
max_pca = coords_pca.max(axis=0)
width = max_pca[0] - min_pca[0]
height = max_pca[1] - min_pca[1]

# --- 5. Elipse (já estava ok) ---
t = np.linspace(0, 2 * np.pi, 100)
ellipse = np.column_stack([(width / 2) * np.cos(t), (height / 2) * np.sin(t)])
ellipse_rotated = ellipse @ components + center
elipse_geom = Polygon(ellipse_rotated)
gdf_elipse = gpd.GeoSeries([elipse_geom], crs=bacia.crs)

# --- 6. Retângulo mínimo envolvente orientado ---
# Canto inferior esquerdo e superior direito no sistema PCA
rect_pca = np.array([
    [min_pca[0], min_pca[1]],
    [max_pca[0], min_pca[1]],
    [max_pca[0], max_pca[1]],
    [min_pca[0], max_pca[1]],
    [min_pca[0], min_pca[1]]
])

# Rotaciona de volta para sistema original
rect_coords = rect_pca @ components + center
retangulo_geom = Polygon(rect_coords)
gdf_retangulo = gpd.GeoSeries([retangulo_geom], crs=bacia.crs)

# --- 7. Centroide para textos ---
centroide = bacia.geometry.centroid.unary_union
centro_x, centro_y = centroide.x, centroide.y
desloc_y = 0

# --- 8. Área do retângulo PCA ---
area_retangulo_m2 = width * height
area_retangulo_km2 = area_retangulo_m2 / 1e6

# --- 9. Textos para os gráficos ---
texto_Kc = (
    r"$K_c = \frac{P}{2 \sqrt{\pi A}}$" + "\n\n"
    f"P: {perimetro_km:.2f} km\n"
    f"A: {area_km2:.2f} km²\n\n"
    f"Resultado:\nKc = {Kc:.3f}"
)

texto_Ic = (
    r"$I_c = \frac{A}{L^2}$" + "\n\n"
    f"Areal bacia: {area_km2:.2f} km²\n"
    f"L (comprimento axial): {comprimento_km:.2f} km\n"
    f"Área retângulo envolvente: {area_retangulo_km2:.2f} km²\n\n"
    f"Resultado:\nIc = {Ic:.3f}\n"
    f"Classificação: {classificacao}\n\n"
    "→ Ic → 1: bacia circular\n"
    "→ Ic → 0: bacia alongada"
)

# --- 10. MAPAS 1x2 ---
fig, axes = plt.subplots(1, 2, figsize=(20, 14))

# 1. Compacidade com elipse
bacia.plot(ax=axes[0], color='none', edgecolor='green', linewidth=2)
gdf_elipse.plot(ax=axes[0], color='none', edgecolor='orange', linestyle='--', linewidth=2)
axes[0].set_title("Coeficiente de Compacidade", fontsize=18, y=1.02)
axes[0].axis('off')
axes[0].text(
    centro_x, centro_y + desloc_y,
    texto_Kc,
    fontsize=16,
    verticalalignment='center',
    horizontalalignment='center',
    bbox=dict(boxstyle='round,pad=0.7', facecolor='white', alpha=0.9)
)

# 2. Índice de Conformação com retângulo orientado
bacia.plot(ax=axes[1], color='none', edgecolor='purple', linewidth=2)
gdf_retangulo.plot(ax=axes[1], color='none', edgecolor='black', linestyle='--', linewidth=2)
axes[1].set_title("Índice de Conformação", fontsize=18, y=1.02)
axes[1].axis('off')
axes[1].text(
    centro_x, centro_y + desloc_y,
    texto_Ic,
    fontsize=16,
    verticalalignment='center',
    horizontalalignment='center',
    bbox=dict(boxstyle='round,pad=0.7', facecolor='white', alpha=0.9)
)

plt.tight_layout()
plt.show()
```

# **Declividade média estimada:**


```python
import rasterio
import numpy as np
from scipy import ndimage

# Caminho do DEM (mesmo do passo anterior)
caminho_dem = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/filled_dem.tif'

# Abrir o DEM e carregar como array
with rasterio.open(caminho_dem) as dem:
    dem_array = dem.read(1)
    transform = dem.transform
    nodata = dem.nodata

# Substituir nodata por nan para evitar influenciar nos cálculos
dem_array = np.where(dem_array == nodata, np.nan, dem_array)

# Calcular gradiente (Sobel) nas direções x e y para estimar declividade
sobel_x = ndimage.sobel(dem_array, axis=1, mode='constant', cval=np.nan)
sobel_y = ndimage.sobel(dem_array, axis=0, mode='constant', cval=np.nan)

# Magnitude do gradiente = declividade aproximada em "pixels"
declividade_pixels = np.hypot(sobel_x, sobel_y)

# Converter declividade de pixels para unidades reais:
# Para isso, precisamos da resolução espacial do DEM (pixel size)
res_x = transform.a  # tamanho do pixel em x (m)
res_y = -transform.e  # tamanho do pixel em y (m), negativo porque transform.e é negativo em north-up

# Declividade real (m/m)
declividade_real = declividade_pixels / np.hypot(res_x, res_y)

# Filtrar valores válidos (remover nan)
declividade_validos = declividade_real[~np.isnan(declividade_real)]

# Calcular a média da declividade
declividade_media = np.mean(declividade_validos)

print(f"Declividade média estimada: {declividade_media:.6f} m/m")
```

```python
import geopandas as gpd
from shapely.geometry import LineString, MultiLineString, Point
import numpy as np
import networkx as nx
from shapely.ops import linemerge # To merge river segments
import pandas as pd # Needed for pd.isna()
import rasterio # To work with the DEM
from rasterio.transform import rowcol # To convert coords to pixel indices

# --- 0. Path Configurations ---
path_drenagem = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/streams_vector.shp' # Using the corrected shapefile now
path_dem = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/filled_dem.tif'
# !!! NEW: You NEED to provide the path to your watershed (basin) shapefile !!!
path_bacia = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/bacia.shp' # <--- REPLACE WITH YOUR ACTUAL BASIN SHAPEFILE PATH

crs_target = 'EPSG:31982'

# --- 1. Load Data ---
print("1. Loading drainage shapefile...")
try:
    drenagem = gpd.read_file(path_drenagem)
    if drenagem.crs is None or drenagem.crs != crs_target:
        print(f"Reprojecting drainage to {crs_target}...")
        drenagem = drenagem.to_crs(crs_target)
    print("✅ Drainage shapefile loaded and CRS adjusted.")
except Exception as e:
    print(f"❌ Error loading drainage shapefile: {e}")
    exit()

print("1.1. Loading watershed (basin) shapefile...")
bacia_gdf = None
try:
    bacia_gdf = gpd.read_file(path_bacia)
    if bacia_gdf.crs is None or bacia_gdf.crs != crs_target:
        print(f"Reprojecting basin to {crs_target}...")
        bacia_gdf = bacia_gdf.to_crs(crs_target)
    if len(bacia_gdf) > 1:
        bacia_gdf = bacia_gdf.dissolve().geometry.iloc[0] # Ensure single Polygon geometry
    else:
        bacia_gdf = bacia_gdf.geometry.iloc[0]
    print("✅ Basin shapefile loaded and CRS adjusted.")
except Exception as e:
    print(f"❌ Error loading basin shapefile: {e}. Basin boundary checks will be skipped.")

# Load DEM
dem_data = None
dem_transform = None
dem_crs = None
dem_loaded_successfully = False
try:
    with rasterio.open(path_dem) as src:
        dem_transform = src.transform
        dem_data = src.read(1) # Read the first band (elevation)
        dem_crs = src.crs
    dem_loaded_successfully = True
    print(f"✅ DEM loaded. DEM CRS: {dem_crs}")
except Exception as e:
    print(f"❌ Error loading DEM: {e}. Elevation-based checks may be less precise.")

# --- 1.5. Generate Node IDs and Extract Node Elevations ---
print("1.5. Generating Node IDs and extracting elevations for the drainage network...")

node_coords_map = {} # Maps (rounded coordinate tuple) -> node_id
node_id_counter = 0 # Counter for assigning new IDs
precision = 3 # Precision for coordinate rounding (e.g., 3 decimal places = millimeters for UTM)

from_node_ids = []
to_node_ids = []
node_elevations = {} # Maps node_id -> elevation
node_id_to_coords = {} # Maps node_id -> original (unrounded) coordinate tuple

for idx, row in drenagem.iterrows():
    geom = row.geometry
    if not isinstance(geom, LineString):
        if isinstance(geom, MultiLineString):
            geom = max(geom.geoms, key=lambda x: x.length, default=None)
            if geom is None:
                from_node_ids.append(None)
                to_node_ids.append(None)
                continue
        else:
            from_node_ids.append(None)
            to_node_ids.append(None)
            continue

    start_coord_raw = geom.coords[0]
    end_coord_raw = geom.coords[-1]

    def get_or_assign_node_id_and_elevation(coord_raw, counter, coords_map, elevations_dict, id_to_coords_dict, dem_data, dem_transform, precision):
        coord_rounded = (round(coord_raw[0], precision), round(coord_raw[1], precision))

        if coord_rounded not in coords_map:
            current_node_id = counter
            coords_map[coord_rounded] = current_node_id
            id_to_coords_dict[current_node_id] = coord_raw
            counter += 1
            if dem_data is not None:
                r, c = rowcol(dem_transform, coord_raw[0], coord_raw[1])
                if 0 <= r < dem_data.shape[0] and 0 <= c < dem_data.shape[1]:
                    elevations_dict[current_node_id] = dem_data[r, c]
                else:
                    elevations_dict[current_node_id] = np.nan
        else:
            current_node_id = coords_map[coord_rounded]
        return current_node_id, counter

    start_node_id, node_id_counter = get_or_assign_node_id_and_elevation(
        start_coord_raw, node_id_counter, node_coords_map, node_elevations, node_id_to_coords, dem_data, dem_transform, precision
    )
    from_node_ids.append(start_node_id)

    end_node_id, node_id_counter = get_or_assign_node_id_and_elevation(
        end_coord_raw, node_id_counter, node_coords_map, node_elevations, node_id_to_coords, dem_data, dem_transform, precision
    )
    to_node_ids.append(end_node_id)

drenagem['from_node_id'] = from_node_ids
drenagem['to_node_id'] = to_node_ids

drenagem = drenagem.dropna(subset=['from_node_id', 'to_node_id']).copy()

print(f"✅ Node IDs generated. Total unique nodes: {node_id_counter}")
if dem_loaded_successfully:
    print(f"✅ Elevations extracted for {len(node_elevations)} nodes.")

# --- 2. Build the Drainage Network Graph ---
print("2. Building the drainage network graph with NetworkX...")

G = nx.DiGraph()

for idx, row in drenagem.iterrows():
    u = int(row['from_node_id'])
    v = int(row['to_node_id'])
    G.add_edge(u, v,
               geometry=row.geometry,
               ordem=row['STRM_VAL'],
               length=row.geometry.length)

print(f"✅ Drainage graph built with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")

# --- 3. Identify Nodes (Sources, Sinks, Junctions) and Create Nodos_gdf ---
print("3. Identifying node types and creating nodos_gdf...")
nodos_data = []
for node_id in G.nodes():
    in_degree = G.in_degree(node_id)
    out_degree = G.out_degree(node_id)
    node_type = 'intermediate'
    if in_degree == 0 and out_degree == 0:
        node_type = 'isolated'
    elif in_degree == 0:
        node_type = 'source' # Nascente
    elif out_degree == 0:
        node_type = 'sink' # Sumidouro / Foz
    elif in_degree > 1 and out_degree >= 1: # Junction where multiple streams merge
        node_type = 'junction'

    elevation = node_elevations.get(node_id, np.nan) if dem_loaded_successfully else np.nan
    coords = node_id_to_coords.get(node_id, (np.nan, np.nan))

    nodos_data.append({
        'node_id': node_id,
        'geometry': Point(coords),
        'in_degree': in_degree,
        'out_degree': out_degree,
        'node_type': node_type,
        'elevacao': elevation
    })

nodos_gdf = gpd.GeoDataFrame(nodos_data, geometry='geometry', crs=crs_target)
print(f"✅ Nodos_gdf created with {len(nodos_gdf)} nodes.")
print(f"Node types summary:\n{nodos_gdf['node_type'].value_counts()}")

# --- 4. Identify the Exutório (Lowest Node on Basin Boundary/Sink) ---
print("4. Identifying the Exutório (Mouth) node...")
exutorio_node = None

# Candidates:
# 1. Nodes classified as 'sink' (out_degree == 0)
# 2. Nodes near the basin boundary (if basin_gdf is available)

potential_exutorio_candidates = nodos_gdf[nodos_gdf['node_type'] == 'sink'].copy()

# Add nodes on basin boundary if basin_gdf is available (and if not already a sink)
if bacia_gdf is not None and not bacia_gdf.is_empty:
    print("   Filtering potential exutorios by basin boundary intersection...")
    # Find nodes that intersect the basin's boundary
    # A small buffer for the node point handles slight misalignments
    nodes_on_boundary_idx = nodos_gdf.geometry.intersects(bacia_gdf.boundary.buffer(1)) # 1m buffer tolerance
    nodes_on_boundary = nodos_gdf[nodes_on_boundary_idx].copy()

    # Add these to candidates, avoiding duplicates (already sinks or already added)
    # Ensure they have *some* outflow (not 'source' or 'isolated' type) if possible
    boundary_outflow_candidates = nodes_on_boundary[
        (nodes_on_boundary['node_type'] != 'source') & (nodes_on_boundary['node_type'] != 'isolated')
    ].copy()

    potential_exutorio_candidates = pd.concat([potential_exutorio_candidates, boundary_outflow_candidates]).drop_duplicates(subset=['node_id'])

    if potential_exutorio_candidates.empty:
        print("   ⚠️ No nodes found as sinks or on basin boundary. Exutório identification difficult.")
else:
    print("   ⚠️ Basin shapefile not available/valid. Relying solely on topological sinks for Exutório candidates.")

if not potential_exutorio_candidates.empty and dem_loaded_successfully and node_elevations:
    # From potential candidates, pick the one with the lowest elevation
    exutorio_candidates_with_elev = potential_exutorio_candidates.dropna(subset=['elevacao'])
    if not exutorio_candidates_with_elev.empty:
        exutorio_node_row = exutorio_candidates_with_elev.sort_values('elevacao', ascending=True).iloc[0]
        exutorio_node = exutorio_node_row['node_id']
        print(f"✅ Exutório node identified by lowest elevation among candidates: {exutorio_node} (Elev: {exutorio_node_row['elevacao']:.2f}m)")
    else:
        print("❌ Potential exutorios found, but none with valid elevation data. Cannot use DEM for final selection.")
elif not potential_exutorio_candidates.empty:
    # Fallback: if no DEM, take the lowest ID sink or most central
    print("⚠️ DEM not available or no valid elevations for candidates. Selecting Exutório heuristically (e.g., first sink).")
    exutorio_node = potential_exutorio_candidates.iloc[0]['node_id'] # Take the first one, NOT ROBUST
    print(f"Exutório node identified by first available candidate: {exutorio_node}")
else:
    print("❌ No valid Exutório candidates found after all checks. Cannot identify Exutório.")

# --- 5. Find the Main River Course (Most Distant Source to Exutório) ---
curso_principal_geom = None

if exutorio_node is not None:
    print("5. Finding the main river course (most distant source to exutório)...")

    # Filter for valid source nodes (in_degree == 0) that have a path to the exutorio
    source_nodes_gdf = nodos_gdf[nodos_gdf['node_type'] == 'source'].copy()

    if source_nodes_gdf.empty:
        print("❌ No source nodes (in_degree == 0) found in the network. Cannot identify main course.")
    else:
        best_path_nodes = None
        max_path_length_stream = -1.0 # Use this to select the longest path in stream length

        # Iterate over all sources and find the longest path to the exutorio
        print(f"   Checking paths from {len(source_nodes_gdf)} source nodes to Exutório {exutorio_node}...")

        # Consider a subset of sources for performance if many, e.g., highest elevation sources
        # source_nodes_to_check = source_nodes_gdf.sort_values('elevacao', ascending=False).head(10)['node_id'].tolist()
        source_nodes_to_check = source_nodes_gdf['node_id'].tolist() # Check all by default

        for source in source_nodes_to_check:
            if nx.has_path(G, source, exutorio_node):
                # Using all_simple_paths can be computationally expensive for very large networks
                # If it's slow, consider limiting the search depth or using a different algorithm
                try:
                    for path in nx.all_simple_paths(G, source, exutorio_node):
                        current_path_length = sum(G[path[i]][path[i+1]]['length'] for i in range(len(path) - 1))

                        if current_path_length > max_path_length_stream:
                            max_path_length_stream = current_path_length
                            best_path_nodes = path
                except Exception as e:
                    print(f"      Warning: Error finding paths from source {source}: {e}. Skipping this source.")

        if best_path_nodes:
            print(f"✅ Main river course identified (longest stream path):")
            print(f"   Start (most distant source): Node {best_path_nodes[0]}")
            print(f"   End (exutório): Node {best_path_nodes[-1]}")
            print(f"   Number of segments: {len(best_path_nodes)-1}")
            print(f"   Real stream length (sum of segments): {max_path_length_stream:.2f} m.")

            # Reconstruct the geometry of the main river course
            segments_in_main_path = []
            for i in range(len(best_path_nodes) - 1):
                u, v = best_path_nodes[i], best_path_nodes[i+1]
                if G.has_edge(u, v):
                    segments_in_main_path.append(G[u][v]['geometry'])

            if segments_in_main_path:
                curso_principal_geom = linemerge(segments_in_main_path)
                if not isinstance(curso_principal_geom, LineString):
                    if isinstance(curso_principal_geom, MultiLineString):
                        print("⚠️ linemerge resulted in MultiLineString for the main course. Taking the longest sub-segment.")
                        curso_principal_geom = max(curso_principal_geom.geoms, key=lambda x: x.length)
                    else:
                        print("⚠️ Unexpected geometry after linemerge for the main course.")
                        curso_principal_geom = None
            else:
                print("Could not reconstruct geometry for the main river course from identified segments.")
                curso_principal_geom = None
        else:
            print("❌ No path found from any source to the identified Exutório with valid length.")
            curso_principal_geom = None

else:
    print("❌ Exutório node not defined. Cannot identify the main course.")

# --- 6. Calculate Sinuosity (Real Length / Axial Length) ---
if curso_principal_geom is None:
    print("❌ Could not calculate sinuosity, as the main course geometry was not defined.")
else:
    print("6. Calculating sinuosity (Real Length / Axial Length)...")

    comprimento_real = curso_principal_geom.length

    if len(curso_principal_geom.coords) < 2:
        print("⚠️ The main course geometry has less than 2 points. Not a valid line for sinuosity.")
        sinuosidade = np.nan
    else:
        nascente_coord = curso_principal_geom.coords[0] # Coordinate of the most distant source
        exutorio_coord = curso_principal_geom.coords[-1] # Coordinate of the exutório

        comprimento_axial = LineString([nascente_coord, exutorio_coord]).length

        print(f"   Real length of the main course: {comprimento_real:.2f} m.")
        print(f"   Axial length (source to mouth): {comprimento_axial:.2f} m.")

        if comprimento_axial > 0:
            sinuosidade = comprimento_real / comprimento_axial
            print(f"✅ Sinuosity of the main course: {sinuosidade:.3f}")
        else:
            print("⚠️ Axial length is zero. The main course forms a closed loop or is a single point.")
            sinuosidade = np.nan

    # Optional: Save the identified main course as a shapefile
    output_main_course_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/curso_principal_completo.shp"
    try:
        gpd.GeoDataFrame(geometry=[curso_principal_geom], crs=crs_target).to_file(output_main_course_path)
        print(f"✅ Complete main course identified and saved as '{output_main_course_path}'")
    except Exception as e:
        print(f"❌ Error saving main course shapefile: {e}")

        # --- 7. Clipar o curso principal pela bacia e salvar ---
print("7. Clipando o curso principal dentro da bacia...")

try:
    # Reabra o shapefile salvo (opcional, se não estiver em memória)
    curso_principal_gdf = gpd.GeoDataFrame(geometry=[curso_principal_geom], crs=crs_target)

    # Clipping
    curso_principal_clipado = gpd.overlay(curso_principal_gdf, gpd.GeoDataFrame(geometry=[bacia_gdf], crs=crs_target), how='intersection')

    # Verificação
    if curso_principal_clipado.empty:
        print("⚠️ O curso principal não intersecta com a bacia. Nenhuma linha clipada.")
    else:
        # Salvar o resultado
        output_clipped_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/curso_principal_clipado.shp"
        curso_principal_clipado.to_file(output_clipped_path)
        print(f"✅ Curso principal clipado salvo em: {output_clipped_path}")
except Exception as e:
    print(f"❌ Erro ao clipar o curso principal: {e}")

```

# **Calcula o tempo de concentração para kirpich**


Aqui será calculado o tempo de concentração para kirpich, porém no caso da bacia do rio da várzea, a bacia não é pequena, então não será colocado no trabalho escrito.


```python
import geopandas as gpd
from shapely.geometry import LineString, Point
import rasterio
import numpy as np

# --- 1. Carregar linha do rio principal ---
rio_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/curso_principal_clipado.shp"
rio = gpd.read_file(rio_path).to_crs("EPSG:31982")

# Unifica as linhas se necessário
linha = rio.geometry.unary_union
if linha.geom_type == "MultiLineString":
    linha = max(linha.geoms, key=lambda x: x.length)

# Extrai os pontos ao longo da linha
coords = list(linha.coords)

# --- 2. Abrir o raster clipado ---
clipped_dem_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/filled_dem_clipado.tif"
with rasterio.open(clipped_dem_path) as src:
    # Extrair elevação para cada ponto da linha
    elevs = [val[0] for val in src.sample(coords)]

# --- 3. Identificar foz e nascente ---
# Foz: ponto com menor elevação
idx_foz = int(np.argmin(elevs))
foz = Point(coords[idx_foz])
foz_elev = elevs[idx_foz]

# Nascente: ponto mais distante da foz ao longo da linha
dists = [linha.project(Point(p)) for p in coords]
dist_foz = dists[idx_foz]
dist_relative = [abs(d - dist_foz) for d in dists]
idx_nascente = int(np.argmax(dist_relative))
nascente = Point(coords[idx_nascente])
nascente_elev = elevs[idx_nascente]

# --- 4. Calcular parâmetros ---
desnivel = nascente_elev - foz_elev
comprimento = abs(dist_foz - dists[idx_nascente])
declividade_media = desnivel / comprimento

# --- 5. Tempo de concentração (Kirpich) ---
Tc = 0.0195 * (comprimento ** 0.77) * (declividade_media ** -0.385)

# --- 6. Mostrar resultados ---
print(f"Nascente: {nascente_elev:.2f} m | Foz: {foz_elev:.2f} m")
print(f"Desnível: {desnivel:.2f} m")
print(f"Comprimento do rio: {comprimento:.2f} m")
print(f"Declividade média: {declividade_media:.4f} m/m")
print(f"Tempo de concentração (Kirpich): {Tc:.2f} minutos")

```

```python
import geopandas as gpd
from shapely.geometry import LineString, Point
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Carregar dados ---
rio_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/curso_principal_clipado.shp"
bacia_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/bacia.shp"
rio = gpd.read_file(rio_path).to_crs("EPSG:31982")
bacia = gpd.read_file(bacia_path).to_crs("EPSG:31982")

# DEM raster
clipped_dem_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/filled_dem_clipado.tif"
with rasterio.open(clipped_dem_path) as src:
    dem = src.read(1)
    transform = src.transform

# Mapa CN raster (médio) - filtrando -9999 e NoData
cn_map_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/mapa_cn.tif"
with rasterio.open(cn_map_path) as src_cn:
    cn_data = src_cn.read(1)
    nodata_val = src_cn.nodata
    mask = (cn_data != nodata_val) & (cn_data != -9999) & (~np.isnan(cn_data))
    cn_validos = cn_data[mask]
    CN = cn_validos.mean()
print(f"CN médio = {CN:.2f}")

# --- 2. Processar linha do rio ---
linha = rio.geometry.unary_union
if linha.geom_type == "MultiLineString":
    linha = max(linha.geoms, key=lambda x: x.length)

coords = list(linha.coords)

with rasterio.open(clipped_dem_path) as src:
    elevs = [val[0] for val in src.sample(coords)]

# --- Corrigir detecção da foz: ignorar valores inválidos (-9999, NaN) ---
elevs_array = np.array(elevs)
mask_valid = (elevs_array != -9999) & (~np.isnan(elevs_array))
coords_valid = [pt for pt, valid in zip(coords, mask_valid) if valid]
elevs_valid = [e for e, valid in zip(elevs, mask_valid) if valid]

# Encontrar índice do menor valor válido
idx_min = np.argmin(elevs_valid)
foz_coord = coords_valid[idx_min]
foz = Point(foz_coord)
foz_elev = elevs_valid[idx_min]

# Calcular distâncias ao longo da linha
dists = [linha.project(Point(p)) for p in coords]

# Distância da foz na linha (usando índice do ponto válido)
dist_foz = dists[coords.index(foz_coord)]

# Distâncias relativas à foz
dist_relative = [abs(d - dist_foz) for d in dists]

# Índice da nascente é o ponto mais distante da foz
idx_nascente = int(np.argmax(dist_relative))
nascente = Point(coords[idx_nascente])
nascente_elev = elevs[idx_nascente]

# --- 3. Parâmetros principais ---
L = abs(dists[idx_nascente] - dist_foz)     # m
ΔH = nascente_elev - foz_elev               # m
S = ΔH / (L/1000)                                  # m/m
A = bacia.geometry.area.sum() / 1e6         # km²
L_km = L / 1000                             # km

# --- 4. Cálculo dos Tempos de Concentração ---
# Picking (min)
Tc_picking = 51.79 * ((L_km ** 2)/S **0.5) **0.333

# Giandotti (h) - com raiz de A
Tc_giandotti = (4 * (A ** 0.5) + 1.5 * L_km) / (0.8 * (ΔH ** 0.5))

# SCS (min) usando CN médio
Tc_scs = 3.42 * (L_km ** 0.8) * ((1000 / CN) - 9) ** 0.7 * (S ** -0.5) * 60

# --- 5. Exibir resultados ---
print("\n📏 Parâmetros da bacia:")
print(f"✅ Nascente: {nascente_elev:.2f} m | Foz: {foz_elev:.2f} m")
print(f"✅ Comprimento do rio: {L_km:.2f} km")
print(f"✅ Desnível: {ΔH:.2f} m | Declividade média: {S:.4f} m/m")
print(f"✅ Área da bacia: {A:.2f} km²")
print(f"✅ CN médio: {CN:.2f}")

print(f"\n🕒 Tempo de Concentração:")
print(f" • Picking      = {Tc_picking:.2f} minutos ({Tc_picking/60:.2f} horas)")
print(f" • Giandotti    = {Tc_giandotti:.2f} horas")
print(f" • SCS (NRCS)   = {Tc_scs:.2f} minutos ({Tc_scs/60:.2f} horas)")

# --- 6. Isócronas ---
tempos_de_concentracao = {
    'Picking': Tc_picking,
    'Giandotti': Tc_giandotti * 60,  # h → min
    'SCS (NRCS)': Tc_scs
}

fig, axs = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)
plt.suptitle("Isócronas de tempo de viagem até a foz por diferentes métodos", fontsize=16)

for i, (metodo, Tc_minutos) in enumerate(tempos_de_concentracao.items()):
    ax = axs[i]
    n_isocronas = 10
    dist_total = linha.length

    isocronas = []
    tempos = []
    for j in range(1, n_isocronas + 1):
        proporcao = j / n_isocronas
        dist = proporcao * dist_total
        pt = linha.interpolate(dist)
        isocronas.append(pt)
        tempos.append(round(proporcao * Tc_minutos, 1))

    bacia.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=1)
    rio.plot(ax=ax, color='blue', linewidth=2)

    for pt, tempo in zip(isocronas, tempos):
        ax.plot(pt.x, pt.y, 'ro')
        ax.text(pt.x, pt.y, f'{tempo} min', fontsize=8, ha='left', va='bottom')

    ax.plot(nascente.x, nascente.y, 'go', label='Nascente')
    ax.plot(foz.x, foz.y, 'ko', label='Foz')

    ax.set_title(f"{metodo} (Tc = {Tc_minutos:.2f} min)")
    ax.set_xlabel("Coord. X (m)")
    ax.set_ylabel("Coord. Y (m)")
    ax.legend(loc='upper right')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
from skimage.graph import MCP_Geometric
import matplotlib.colors as mcolors
from rasterio.features import geometry_mask


# === 1. Caminhos ===
rio_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/curso_principal_clipado.shp"
bacia_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/bacia.shp"
rede_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/ordem_strahler_clip_lines.shp"
dem_path = "/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/filled_dem_clipado.tif"

# === 2. Carregar dados ===
rio = gpd.read_file(rio_path).to_crs("EPSG:31982")
bacia = gpd.read_file(bacia_path).to_crs("EPSG:31982")
rede = gpd.read_file(rede_path).to_crs("EPSG:31982")

with rasterio.open(dem_path) as src:
    dem = src.read(1, masked=True)
    transform = src.transform
    profile = src.profile
    crs = src.crs
    width, height = src.width, src.height

# === 3. Encontrar foz e nascente no curso principal ===
linha = rio.geometry.unary_union
if linha.geom_type == "MultiLineString":
    linha = max(linha.geoms, key=lambda x: x.length)

coords = list(linha.coords)
with rasterio.open(dem_path) as src:
    elevs = [val[0] for val in src.sample(coords)]

elevs_array = np.array(elevs)
mask_valid = (elevs_array != -9999) & (~np.isnan(elevs_array))
coords_valid = [pt for pt, valid in zip(coords, mask_valid) if valid]
elevs_valid = [e for e, valid in zip(elevs, mask_valid) if valid]

idx_min = np.argmin(elevs_valid)
foz_coord = coords_valid[idx_min]
foz_x, foz_y = foz_coord
foz_idx = ~np.isnan(dem) & (dem.data == dem.data)  # placeholder

dists = [linha.project(Point(p)) for p in coords]
dist_foz = dists[coords.index(foz_coord)]
dist_relative = [abs(d - dist_foz) for d in dists]
idx_nascente = int(np.argmax(dist_relative))
nascente = Point(coords[idx_nascente])
nascente_elev = elevs[idx_nascente]

# Parâmetros
L = abs(dists[idx_nascente] - dist_foz)     # m
ΔH = nascente_elev - elevs_valid[idx_min]   # m
S = ΔH / (L/1000)                           # m/m
A = bacia.geometry.area.sum() / 1e6         # km²
L_km = L / 1000                             # km

# === 4. Calcular Tc ===
Tc_picking = 51.79 * ((L_km ** 2) / (S ** 0.5)) ** 0.333
Tc_giandotti = (4 * (A ** 0.5) + 1.5 * L_km) / (0.8 * (ΔH ** 0.5)) * 60
CN = 76.83  # ajuste se tiver valor médio real
Tc_scs = 3.42 * (L_km ** 0.8) * ((1000 / CN) - 9) ** 0.7 * (S ** -0.5) * 60

tempos_de_concentracao = {
    'Picking': Tc_picking,
    'Giandotti': Tc_giandotti,
    'SCS (NRCS)': Tc_scs
}

# Criar máscara da bacia no mesmo grid do DEM
mask_bacia = geometry_mask(
    [geom for geom in bacia.geometry],
    transform=transform,
    invert=True,
    out_shape=(height, width)
)


def gerar_isocronas(Tc_minutos):
    v_media = L / Tc_minutos
    gy, gx = np.gradient(dem.filled(np.nan))
    slope = np.hypot(gx, gy)
    slope = np.where(slope > 0, slope, 0.001)
    alpha = 0.5
    slope_med = np.nanmean(slope)
    v_local = v_media * (slope / slope_med) ** alpha
    pixel_size_x = abs(transform.a)
    pixel_size_y = abs(transform.e)
    dist_pixel = np.sqrt(pixel_size_x**2 + pixel_size_y**2)

    # Aqui o custo deve ser o tempo para atravessar o pixel, que é inversamente proporcional à velocidade:
    tempo_pixel = dist_pixel / v_local  # tempo por pixel, correto

    # Garante que não existam zeros nem valores inválidos
    tempo_pixel = np.where(tempo_pixel <= 0, np.nanmax(tempo_pixel)*10, tempo_pixel)
    tempo_pixel = np.nan_to_num(tempo_pixel, nan=np.nanmax(tempo_pixel)*10)

    row, col = ~transform * (foz_x, foz_y)
    row, col = int(row), int(col)
    mcp = MCP_Geometric(tempo_pixel)
    costs, _ = mcp.find_costs([(row, col)])

    costs = np.where(mask_bacia, costs, np.nan)
    return costs

```

```python
import geopandas as gpd
import rasterio
import rasterio.mask
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import sobel
from shapely.geometry import mapping

# --- 1. Caminhos ---
shapefile_path = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/bacia.shp'
dem_path = '/content/drive/MyDrive/hidrologia/SHAPEFILES_CRIADOS/filled_dem.tif'

# --- 2. Carregar bacia ---
bacia = gpd.read_file(shapefile_path)

# --- 3. Abrir DEM e cortar pela bacia ---
with rasterio.open(dem_path) as src:
    bacia = bacia.to_crs(src.crs)
    geoms = [mapping(geom) for geom in bacia.geometry]
    out_image, out_transform = rasterio.mask.mask(src, geoms, crop=True)
    elev = out_image[0]
    elev[elev == src.nodata] = np.nan
    res_x, res_y = src.res

# --- 4. Calcular elevação média ---
elev_mean = np.nanmean(elev)

# --- 5. Calcular declividade ---
dx = sobel(elev, axis=1, mode='constant')
dy = sobel(elev, axis=0, mode='constant')
slope_rad = np.arctan(np.sqrt(dx**2 + dy**2) / res_x)
slope_deg = np.degrees(slope_rad)
slope_percent = np.tan(slope_rad) * 100

slope_mean_deg = np.nanmean(slope_deg)
slope_mean_pct = np.nanmean(slope_percent)

# --- 6. Centroide da bacia ---
centroide = bacia.geometry.centroid.unary_union
centro_x, centro_y = centroide.x, centroide.y

# --- 7. Calcular extent para plotar ---
extent = (
    out_transform[2],  # xmin
    out_transform[2] + elev.shape[1] * out_transform[0],  # xmax
    out_transform[5] + elev.shape[0] * out_transform[4],  # ymin
    out_transform[5]   # ymax
)

# --- 8. Plotar resultados ---
fig, axes = plt.subplots(1, 2, figsize=(16, 8), constrained_layout=True)

# 1. Mapa de Elevação
elev_img = axes[0].imshow(elev, cmap='terrain', origin='upper', extent=extent)
bacia.boundary.plot(ax=axes[0], color='black', linewidth=1)
axes[0].set_title("Elevação (m)", fontsize=16)
fig.colorbar(elev_img, ax=axes[0], label='m', orientation='horizontal', fraction=0.046, pad=0.04)
axes[0].axis('on')
axes[0].text(
    0.5, -0.15,
    f"Elevação média: {elev_mean:.2f} m",
    transform=axes[0].transAxes,
    fontsize=12,
    ha='center'
)

# 2. Mapa de Declividade
slope_img = axes[1].imshow(slope_deg, cmap='viridis', origin='upper', extent=extent)
bacia.boundary.plot(ax=axes[1], color='black', linewidth=1)
axes[1].set_title("Declividade (graus)", fontsize=16)
fig.colorbar(slope_img, ax=axes[1], label='graus', orientation='horizontal', fraction=0.046, pad=0.04)
axes[1].axis('on')
axes[1].text(
    0.5, -0.15,
    f"Declividade média: {slope_mean_deg:.2f}° / {slope_mean_pct:.2f}%",
    transform=axes[1].transAxes,
    fontsize=12,
    ha='center'
)

plt.show()

# --- 9. Imprimir no console também ---
print("🌄 Elevação média:", round(elev_mean, 2), "m")
print("📈 Declividade média:", round(slope_mean_deg, 2), "graus")
print("📈 Declividade média:", round(slope_mean_pct, 2), "%")
```
